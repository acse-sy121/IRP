```python
import numpy as np
import matplotlib.pyplot as plt
```

# Scenario 1

```python
Pred_Human = np.load('/content/drive/MyDrive/output_files/DATA/Human_prediction/Case_1/Pred_data.npy')
CFD_Human = np.load('/content/drive/MyDrive/output_files/DATA/Human_prediction/Case_1/CFD_data.npy')
```

```python
x_axis = np.linspace(5, 1800, 360)
```

```python
fig, ax = plt.subplots(9, 3, figsize=[18,20])
fig.tight_layout(h_pad=3.5, w_pad=3.5)
for i in range(27):
  ax.flatten()[i].plot(x_axis, Pred_Human[:,0,i], '-')
  ax.flatten()[i].plot(x_axis, CFD_Human[:,0,i]*1e6, '-')
  ax.flatten()[i].legend(['Pred_Person', 'Real_Person'])
  ax.flatten()[i].set_title('Person '+ str(i+1))
  ax.flatten()[i].set_xlabel('Time (s)')
  ax.flatten()[i].set_ylabel('CO2_ppm')
path = '/content/drive/MyDrive/output_files/Figures/Human_Prediction/Case_1/Tracer.png'
plt.savefig(path)
```

```python
fig, ax = plt.subplots(9, 3, figsize=[18,20])
fig.tight_layout(h_pad=3.5, w_pad=3.5)
for i in range(27):
  ax.flatten()[i].plot(x_axis, Pred_Human[:,1,i], '-')
  ax.flatten()[i].plot(x_axis, CFD_Human[:,1,i], '-')
  ax.flatten()[i].legend(['Pred_Person', 'Real_Person'])
  ax.flatten()[i].set_title('Person '+ str(i+1))
  ax.flatten()[i].set_xlabel('Time (s)')
  ax.flatten()[i].set_ylabel('Viral load')
path = '/content/drive/MyDrive/output_files/Figures/Human_Prediction/Case_1/Viral_load.png'
plt.savefig(path)
```
